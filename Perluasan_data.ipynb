```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import *
from selenium.webdriver.common.keys import Keys
import pandas as pd
import sys
import time

sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')
```

**initial dataset kamus lexicon**

```python
df = pd.read_excel("kamus.xlsx")
df.head()
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument("user-agent=Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36")
# chrome_options.add_argument("user-data-dir=cache")
browser = webdriver.Chrome('chromedriver',options=chrome_options)

browser.get("https://www.sinonimkata.com/sinonim-22618-indonesia.html")

for j in df['Indonesian (id)']:
  wait = WebDriverWait(browser,10,poll_frequency=1,ignored_exceptions=[ElementNotVisibleException, ElementNotSelectableException,NoSuchElementException])
  element = wait.until(EC.element_to_be_clickable((By.XPATH, "//div")))
  search_box = element.find_element(By.NAME,"q")
  search_box.clear()
  search_box.send_keys(j)
  search_box.send_keys(Keys.RETURN)

  lemmas = browser.find_element(By.CLASS_NAME,"lemmas")
  elm = lemmas.find_elements(By.TAG_NAME,"a")
  for i in elm:
    print(i.text)
browser.quit()
```
